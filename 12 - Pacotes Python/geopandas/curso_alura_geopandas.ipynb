# Geopandas: trabalhando com dados geoespaciais

## Introdução ao GeoPandas

```python
import geopandas as gpd
```

```python
url = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/RJ_Municipios_2022.shp'

rj = gpd.read_file(url)
```

```python
rj.head()
```

```python
type(rj)
```

```python
rj.plot(figsize=(10,10))
```

```python
rj.plot(color='white',edgecolor='red',figsize=(10,10))
```

```python
rj_cidade = rj[rj['NM_MUN']=='Rio de Janeiro']
```

```python
rj_cidade
```

```python
rj_cidade.plot(color='orange',figsize=(10,10))
```

```python
rj_cidade.to_file('rj_cidade.shp')
```

```python
gpd.read_file('rj_cidade.shp')
```
