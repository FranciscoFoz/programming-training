# REQUEST JSON API - GITHUB

```python
from urllib.request import urlopen, Request
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.options.display.max_columns = 999
```

## Página

```python
pagina = 'https://api.github.com/users/FranciscoFoz/repos'
```

# Funções

```python
def resposta_json(url):
    '''
    Esta função recebe uma URL como entrada, faz uma solicitação HTTP usando o cabeçalho armazenado na variável de instância headers, 
    lê a resposta HTTP e retorna o JSON da resposta.
    '''
    try: 
        req = Request(url)
        response = urlopen(req)
        json = response.read()
    except HTTPError as e:
        print(e.status, e.reason)
    except URLError as e:
        print(e.reason)
    return json
```

```python
francisco_foz_repos_github = resposta_json(pagina)
json_francisco_foz_repos_github = json.loads(francisco_foz_repos_github)
json_francisco_foz_repos_github
```

```python
df_francisco_foz_repos_github = pd.DataFrame(json_francisco_foz_repos_github)
df_francisco_foz_repos_github.head()
```

```python
repos_data_linguagem = df_francisco_foz_repos_github[['created_at','language']]

repos_data_linguagem['ano'] = pd.to_datetime(repos_data_linguagem.created_at).dt.year
repos_data_linguagem['mes'] = pd.to_datetime(repos_data_linguagem.created_at).dt.month
```

```python
repos_data_linguagem[['ano','language']].value_counts().reset_index().sort_values(by='ano')
```

```python
repos_data_linguagem[['mes','language']].value_counts().reset_index().sort_values(by='mes')
```
