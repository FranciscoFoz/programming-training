## Trabalhando com sistemas de coordenadas

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/curso_geopandas/main/dados/dados.csv',sep='\t')
df
```

```python
list(zip(df['Longitude'],df['Latitude']))
```

```python
from shapely.geometry import Point

```

```python
pontos_geometricos = [zip(df['Longitude'],df['Latitude'])]
```

```python
pontos_geometricos = [Point(x) for x in zip(df['Longitude'],df['Latitude'])]
```

```python
crs = {'proj': 'latlong', 'ellps':'WGS84', 'datum': 'WGS84','no_defs':True}
geo_dados = gpd.GeoDataFrame(df, crs=crs, geometry=pontos_geometricos)
```

```python
geo_dados
```

```python
geo_dados.plot(alpha=0.2,figsize=(10,10))
```

```python
rj.crs
```

```python
geo_dados.crs
```

Que zona UTM em que o município do Rio de Janeiro está inserido?

    O município do Rio de Janeiro está localizado na Zona UTM (Universal Transverse Mercator) 23S. A Zona UTM 23S abrange uma parte do Brasil, incluindo o estado do Rio de Janeiro, e é uma das divisões do sistema de projeção UTM, que é utilizado para representar a superfície da Terra em mapas.

Qual o ESPG para zona UTM 23S utilizando o SIRGAS 2000?

```python
rj_crs = rj.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
```

```python
url_t = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/trem.geojson'
url_b = 'https://github.com/alura-cursos/curso_geopandas/raw/main/dados/brt.geojson'
```

```python
trem = gpd.read_file(url_t)
brt = gpd.read_file(url_b)

trem = trem.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
brt = brt.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
```

```python
trem = trem[trem.within(rj_cidade.iloc[0].geometry)]
brt = brt[brt.within(rj_cidade.iloc[0].geometry)]
```

```python
base = rj_cidade.plot(color='white',edgecolor='black',figsize=(15,10))
geo_dados_crs.plot(ax=base,alpha=0.2)
metro.plot(ax=base, color='#b60a1c',markersize=30)
trem.plot(ax=base, color='#e39802',markersize=30)
brt.plot(ax=base, color='#309143',markersize=30)
```

```python
transporte = pd.concat([trem.geometry,metro.geometry,brt.geometry],ignore_index = True)
```

```python
geo_dados_crs['Dist_Transporte'] = geo_dados_crs['geometry'].apply(lambda x: transporte.distance(x).min())
```

```python
import matplotlib.pyplot as plt


correlacao = geo_dados_crs.corr()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='RdYlBu', linewidth=0.5, ax=ax)
plt.show()
```
